```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

'''
dalam cell ini, kalian akan melakukan import dari library yang sudah kalian install.
kita akan menggunakan pandas sebagai pembaca data,
train_test_split untuk membagi data latihan dan data uji coba dari file train.csv,
LinearRegression adalah nama model ML yang kita akan guankan disini
mean_squared_error dan mean_absolute_error adalah metrik yang akan kita gunakan
untuk mengukur performa model kita, simpelnya selisih hasil prediksi dengan rata rata nilai aslinya.
'''
```

```python
df = pd.read_csv('train.csv')
print(df.head())

'''
kita menggunakan pandas(pd) di sini untuk membaca file train.csv 
lalu memanggil train.csv untuk dilihat sekilas.
df.head() akan memunculkan 5 baris data teratas dari file train.csv
df.tail() akan memunculkan 5 baris data terbawah dari file train.csv
df.head(10)...? coba sendiri aja, hehe.
'''
```

```python
model = LinearRegression()
X = df.drop('harga', axis=1)
y = df['harga']

'''
disini kita definisikan model = LinearRegression(), biar simpel aja sebenernya, jadi lebih pendek.
lalu kita definisikan X sebagai fitur yang digunakan untuk memprediksi,
dan y adalah hasil yang akan kita prediksi/target prediksi
'''
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 4211)
model.fit(X_train, y_train)
MAE_train = (mean_absolute_error(y_train, model.predict(X_train)))
RMSE_train = (mean_squared_error(y_train, model.predict(X_train)))
X_train.index

'''
pertama kita bagi datanya, yaitu ada X_train, X_test, y_train, y_test.
:param test_size=0.2 artinya 20% data akan digunakan sebagai data uji coba, dan sisanya untuk latihan:
:param random_state=4211 adalah id unik. isinya bebas, tapi akan selalu sama tiap kali dijalankan:
lalu kita fit modelnya menggunakan data latihan, X latihan dan y latihan
lalu kita hitung MAE dan RMSE pada latihan(satu saja cukup)
index digunakan untuk melacak data mana yang digunakan pada saat latihan(opsional)
'''
```
